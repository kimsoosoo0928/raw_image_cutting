```python
import numpy as np
```

```python
fd = open('./data/123_refl.raw','rb')
data = fd.read()
rows = 1536
cols = 2048

blue = np.zeros((rows,cols)).astype(np.uint8)
green = np.zeros((rows,cols)).astype(np.uint8)
red = np.zeros((rows,cols)).astype(np.uint8)
n1 = np.zeros((rows,cols)).astype(np.uint8)
n2 = np.zeros((rows,cols)).astype(np.uint8)

print(len(data))
# band = np.array([blue, green, red, n1, n2])

for row in range(0,rows):
    for col in range(0,cols):
        blue[row,col] = data[row*cols*5+col]
        green[row,col] = data[row*cols*5+col+(cols)]
        red[row,col] = data[row*cols*5+col+(2*cols)]
        n1[row,col] = data[row*cols*5+col+(3*cols)]
        n2[row,col] = data[row*cols*5+col+(4*cols)]



import cv2
cv2.imshow('',blue)
cv2.waitKey()
cv2.destroyAllWindows()
```
